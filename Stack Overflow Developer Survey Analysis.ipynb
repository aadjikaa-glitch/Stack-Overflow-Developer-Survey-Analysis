```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Створено Дата Фрейм
df = pd.read_csv('survey_results_public.csv')
df2 = pd.read_csv('survey_results_schema.csv')
```

```python
df.head()
```

```python
#1. Загальна кількість користувачів, які взяли участь в опитуванні Stack Overflow

count_respons = df['ResponseId'].size
count_respons
```

```python
#Перетворення у список

questions = df2['qname'].tolist()
```

```python
#Отримання списку питань 

stuck = []
for q in questions:
    if q in df.columns:
        stuck.append(q)
questions = stuck 
print(questions)
```

```python
#Перевірка на наявність відповідей на ці питання у кожного респондента
check_all = df[questions].notna().all(axis = 1)
check_all
```

```python
#Кількість респондентів без пропущених значень
count_all = check_all.sum()
count_all
```

```python
# Обчислення не пропущенних значень

non_missing_values = df['WorkExp'].count()
non_missing_values
```

```python
# Обчислення пропущенних значень

missing_values = df['WorkExp'].isna().sum()
missing_values
```

```python
#Середнє значення (mean)
#Медіана (median)
#Мода (mode)

mean_ = df['WorkExp'].mean()
median_ = df['WorkExp'].median()
mode_ = df['WorkExp'].mode()
```

```python
df['WorkExp'].describe()
```

```python
#Виведення результатів

print(mean_)
print(median_)
print(mode_)
```

```python
## Аналіз формату роботи розробників
#За допомогою циклу `for` виконано перебір значень у стовпці та підраховано кількість записів зі значенням `"Remote"`.

rem = 0
for w in df['RemoteWork']:
    if w == 'Remote':
         rem +=1
rem        
```

```python
#Aналіз розподілу форматів роботи респондентів за допомогою методу `value_counts()`, включаючи пропущені значення.

df['RemoteWork'].value_counts(dropna=False)
```

```python
#Перевірка на наявність Python у колонці 'LanguageHaveWorkedWith' та підрахунок кількості респондентів, які програмують на Python

python_users = df['LanguageHaveWorkedWith'].fillna('').str.contains('Python').sum()
python_users
```

```python
# Підраховано кількість непорожніх значень у колонці `LanguageHaveWorkedWith`+
# обчислено відсоток користувачів Python серед усіх респондентів.
total = df['LanguageHaveWorkedWith'].notna().sum()
percentage = (python_users/total)*100
percentage
```

```python
#Перевірка на наявність 'Online Courses or Certification' у колонці 'LearnCode'
#та підрахунок кількості респондентів , які обрали онлайн курси.

courses = df['LearnCode'].fillna('').str.contains('Online Courses or Certification').sum()
courses
```

```python
#Стіорюємо дф, де залишаємо тілки тих респондентів, що використовують Python
python = df['LanguageHaveWorkedWith'].fillna('').str.contains('Python')
```

```python
# створюємо копію 
df_python = df[python].copy()
df_python
```

```python
# Підготовка даних для аналізу заробітної плати
#Для подальшого аналізу рівня заробітної плати значення у колонці `ConvertedCompYearly` було перетворено у числовий формат.
#Параметр `errors='coerce'` використано для автоматичної заміни некоректних або текстових значень на `NaN`, 
#що дозволяє коректно виконувати статистичний аналіз даних.

df_python['ConvertedCompYearly'] = pd.to_numeric(df_python['ConvertedCompYearly'], errors = 'coerce')
```

```python
#Групуємо значення за країнами та підраховуємо 'mean','median'
result = df_python.groupby('Country')['ConvertedCompYearly'].agg(['mean','median'])
result
```

```python
#Обрано колонки з інформацією про рівень освіти та заробітну плату, видалено пропущені значення, 
#виконано сортування даних за спаданням заробітної плати, визначено топ-5 респондентів із найвищим доходом.
    
df_top_5 = df[['EdLevel','ConvertedCompYearly']].dropna().sort_values(by = 'ConvertedCompYearly', ascending = False).head(5)
df_top_5
```
